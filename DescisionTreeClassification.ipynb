```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("social_ads.csv")
print(dataset.head(3))
```

```python
sns.scatterplot(x="Age",y="EstimatedSalary",hue="Purchased",data=dataset)
plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset["Purchased"]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
sc.fit(x)
x=pd.DataFrame(sc.transform(x),columns=x.columns)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier(criterion="entropy")
dt.fit(x_train,y_train)
print(dt.score(x_test,y_test)*100)
print(dt.score(x_train,y_train)*100)
```

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(dt)
plt.show()
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=dt)
plt.show()
```

```python
dt1=DecisionTreeClassifier(criterion="entropy",max_depth=5)
dt1.fit(x_train,y_train)
print(dt1.score(x_test,y_test)*100)
print(dt1.score(x_train,y_train)*100)
```

```python
for i in range(1,20):
    dt2=DecisionTreeClassifier(criterion="entropy",max_depth=i)
    dt2.fit(x_train,y_train)
    print(dt2.score(x_test,y_test)*100,dt2.score(x_train,y_train)*100,i)
```
